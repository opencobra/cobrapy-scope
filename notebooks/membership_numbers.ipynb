```python
%matplotlib inline
```

```python
df = pd.read_csv("../data/cobrapie_members.csv")
```

```python
df["time"] = pd.to_datetime(
    df["Join year"].astype(str) + "-" +
    df["Join month"].astype(str) + "-" +
    df["Join day"].astype(str),
    format="%Y-%m-%d"
)
```

```python
df.set_index("time", drop=False, inplace=True)
```

```python
df.sort_index(inplace=True)
```

```python
df.resample('M').size().plot();
```

```python
df.resample('M').size().cumsum().plot();
```
